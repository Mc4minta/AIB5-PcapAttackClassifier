# Combine and clean

```python
df = pd.concat([cic2017, cic2018])
```

```python
df.replace([np.inf, -np.inf], np.nan, inplace=True)
df.dropna(inplace=True)
df.reset_index(drop=True, inplace=True)
```

```python
print(df.info())
print(df['Label'].value_counts())
```

```python
def apply_custom_label_format(label_value):
    label_value = str(label_value).lower().strip()

    if 'benign' in label_value:
        return 'Benign'
    elif 'ftp' in label_value and ('bruteforce' in label_value or 'patator' in label_value):
        return 'FTP-Bruteforce'
    elif 'ssh' in label_value and ('bruteforce' in label_value or 'patator' in label_value):
        return 'SSH-Bruteforce'
    elif 'portscan' in label_value:
        return 'PortScan'
    elif ('slowhttptest' in label_value or 'slowloris' in label_value) and 'dos' in label_value:
        return 'DoS-Slow-Rate'
    elif ('hulk' in label_value or 'goldeneye' in label_value) and 'dos' in label_value:
        return 'DoS-HTTP-Flood'
    else:
        return 'Other'
```

```python
df['Label'] = df['Label'].apply(apply_custom_label_format)
```

```python
print(df.info())
print(df['Label'].value_counts())
```
