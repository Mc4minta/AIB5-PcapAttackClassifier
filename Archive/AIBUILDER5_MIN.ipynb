```python
import pandas as pd
import os

# Define the directory containing the CSV files
# csv_dir = 'CIC-IDS-2017/CSVs/MachineLearningCSV/MachineLearningCVE'
csv_dir = '/content/drive/MyDrive/Share to Mc4/AIBuilders5-MiN/MachineLearningCVE'

# List of CSV file names
csv_files = [
    'Friday-WorkingHours-Afternoon-DDos.pcap_ISCX.csv',
    'Friday-WorkingHours-Afternoon-PortScan.pcap_ISCX.csv',
    'Friday-WorkingHours-Morning.pcap_ISCX.csv',
    'Monday-WorkingHours.pcap_ISCX.csv',
    'Thursday-WorkingHours-Afternoon-Infilteration.pcap_ISCX.csv',
    'Tuesday-WorkingHours.pcap_ISCX.csv',
    'Wednesday-workingHours.pcap_ISCX.csv',
    'Thursday-WorkingHours-Morning-WebAttacks.pcap_ISCX.csv'
]

# Create an empty list to store the DataFrames
dfs = []

# Loop through each CSV file
for file in csv_files:
    # Create the full file path
    file_path = os.path.join(csv_dir, file)

    # Read the CSV file into a DataFrame
    try:
        df = pd.read_csv(file_path)
        dfs.append(df)
        print(f"Successfully loaded: {file}")
    except FileNotFoundError:
        print(f"File not found: {file}")

# Concatenate all DataFrames into a single DataFrame
TrafficLabelling = pd.concat(dfs, ignore_index=True)
# Rename columns from ' Label' to 'Label'
TrafficLabelling.rename(columns={' Label': 'Label'}, inplace=True)
```

```python
# Groupped DOS Labels and rename FTP and SSH to be bruteforce
label_mapping = {
    'BENIGN': 'Benign',
    'DoS Hulk': 'DoS',
    'DDoS': 'DDoS',
    'DoS GoldenEye': 'DoS',
    'DoS slowloris': 'DoS',
    'DoS Slowhttptest': 'DoS',
    'PortScan' : 'PortScan',
    'FTP-Patator': 'FTP-Bruteforce',
    'SSH-Patator': 'SSH-BruteForce',
    # 'Web Attack � Brute Force	' : 'Web-BruteForce',
    # 'Web Attack � XSS' : 'XSS',
    # 'Web Attack � Sql Injection' : 'SQL-Injection',
}

# Apply the label mapping to the 'Label' column
cic2017['Label'] = cic2017['Label'].map(label_mapping)
cic2017['Label'].value_counts()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
from sklearn.preprocessing import LabelEncoder

# ... (Your existing code for loading and preprocessing the data: loading cic2017, feature selection, etc.) ...

# Assuming 'features' DataFrame and 'cic2017' DataFrame are already loaded

# Select features from 'features' DataFrame
X = features.drop(columns=['Label'])  # Exclude the 'Label' column from features

# Get the label from the 'cic2017' DataFrame (assuming it has the label)
# Ensure that the labels in y are of string type before fitting LabelEncoder
y = features['Label'].astype(str)

# Encode labels if they are strings (e.g., "DDoS", "Normal")
le = LabelEncoder()
y_encoded = le.fit_transform(y)

# Split into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y_encoded, test_size=0.2, random_state=42)

# --------------------- Handling Infinite/Large Values and Non-numerical Values --------------------- #
# Replace infinite values with NaN
X_train = X_train.replace([np.inf, -np.inf], np.nan)
X_test = X_test.replace([np.inf, -np.inf], np.nan)  # Apply to X_test as well

# Convert all columns in X_train and X_test to numeric, coercing errors to NaN
for col in X_train.columns:
    X_train[col] = pd.to_numeric(X_train[col], errors='coerce')
    X_test[col] = pd.to_numeric(X_test[col], errors='coerce')

# Impute NaN values using the median for both X_train and X_test
for col in X_train.columns:
    X_train[col] = X_train[col].fillna(X_train[col].median())
    X_test[col] = X_test[col].fillna(X_test[col].median())  # Impute for X_test using X_train median

# Convert data types to float64 for wider range
X_train = X_train.astype(np.float64)
X_test = X_test.astype(np.float64)  # Convert X_test as well
# --------------------------------------------------------------------------------------------------- #

# Train Random Forest
model = RandomForestClassifier(n_estimators=100, random_state=42)
model.fit(X_train, y_train)

# Predict and evaluate
y_pred = model.predict(X_test)

# Convert the encoded predictions back to original labels for the classification report
y_pred_original = le.inverse_transform(y_pred)
y_test_original = le.inverse_transform(y_test)

print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred))
print("\nClassification Report:\n", classification_report(y_test_original, y_pred_original, target_names=le.classes_))  # Use original labels here
print("\nAccuracy:", accuracy_score(y_test, y_pred))
```

```python
import pandas as pd
import matplotlib.pyplot as plt

def plot_port_vs_others(DstPort, Attack):
    Attack = Attack.strip()  # Remove any leading/trailing spaces from Attack label

    # Count attacks of the specified type ('yes') and others ('no') for the given destination port
    yes = len(cic2017[(cic2017['Label'].str.strip() == Attack) & (cic2017['Destination Port'] == DstPort)])
    no = len(cic2017[(cic2017['Label'].str.strip() != Attack) & (cic2017['Destination Port'] == DstPort)])

    plt.figure(figsize=(8, 6))
    bars = plt.bar([f'Is {Attack}', f'Not {Attack}'], [yes, no], color=['skyblue', 'salmon'])
    plt.title(f'{Attack} Attacks: Destination Port {DstPort} vs. Others')
    plt.ylabel('Number of Attacks')

    total = yes + no  # Calculate total number of attacks for this port
    if total == 0:
        print(f"No attacks of type {Attack} found for destination port {DstPort}.")
        return  # Exit the function if no data is available for the specified attack

    # Display the percentage on top of each bar
    for bar in bars:
        height = bar.get_height()
        percentage = (height / total) * 100
        plt.text(bar.get_x() + bar.get_width() / 2, height + 5, f'{percentage:.1f}%', ha='center', va='bottom')

    plt.show()
```
